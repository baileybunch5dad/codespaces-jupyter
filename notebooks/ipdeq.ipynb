integrating factors

text example



\begin{align}
\frac{dy}{dx} + 3y &= 9 & \text{solve} \\
y(0) &= 7 & \text{initial condition} \\
\mu(x) \frac{dy}{dx} + 3\mu(x) y &= \mu(x) 9 & \text{multiply by } \mu(x) \\
\frac{d}{dx} \left( \mu(x)y \right) &= \mu(x) \frac{d}{dx} \left( y \right)  + \frac{d}{dx} \left(( \mu(x) \right) y & \text{generic chain rule}\\
&= \mu(x) \frac{dy}{dx} + y \frac{d \mu(x)}{dx} \\
& \boxed{ \mu(x) \frac{dy}{dx} + y \frac{d \mu(x)}{dx} } & \text{right hand side of this generic chain rule}\\
& \boxed{ \mu(x)\frac{dy}{dx} + 3\mu(x)y } & \text{ left hand side of diffeq to solve }\\
\mu(x) \frac{dy}{dx} + y \frac{d \mu(x)}{dx} &= \mu(x)\frac{dy}{dx} + 3\mu(x)y  & \text{transforming known to given diffeq} \\
\cancel{\mu(x) \frac{dy}{dx}} + y \frac{d \mu(x)}{dx} &= \cancel{\mu(x)\frac{dy}{dx}} + 3\mu(x)y  & \text{algebra} \\
y \frac{d \mu(x)}{dx} &= 3\mu(x)y \\
\cancel{{y}} \frac{d \mu(x)}{dx} &= 3\mu(x)\cancel{y} \\
\frac{d \mu(x)}{dx} &= 3\mu(x)   \\
\frac{d \mu(x)}{dx}\frac{1}{\mu{x}} &= 3 \\
\frac{d \mu(x)}{\mu(x)} &= 3 dx & \text{separated, now integrate}\\
\int \frac{d \mu(x)}{\mu(x)} &= \int 3 dx \\
\ln|\mu{x}| &= 3x + c \\
e^{\ln|\mu{x}|} &= e^{3x + c} \\
\mu(x) &= ce^{3x} &\text{bingo, found integrating factor }\mu(x) &= ce^{3x}\\
ce^{3x} \frac{dy}{dx} + 3ce^{3x} y &= 9 ce^{3x} \\
e^{3x} \frac{dy}{dx} + 3e^{3x} y &= 9 e^{3x} & c \text{ drops} \\
uv' + u'v &= d(uv) & \text{ lhs is from chain rule, where } u=e^{3x}, v=y \\
\frac{d (e^{3x}y)}{dx} &= 9 e^{3x} \\
d (e^{3x}y) &= 9 e^{3x} dx \\
\int d (e^{3x}y) &= \int 9 e^{3x} dx \\
e^{3x}y &= 3e^{3x} + c \\
\frac{e^{3x}y}{e^{3x}} &= \frac{3e^{3x} + c}{e^{3x}} \\
y &= \frac{3e^{3x}}{e^{3x}} + \frac{c}{e^{3x}} \\
&= \boxed{ 3 + ce^{-3x} } & \text{ solved for } y \\
y(0) &= y = 3 + ce^{-30} & \text{plug in initial condition}\\
c &= 4 \\
y &= \boxed{ 3 + 4e^{-3x}} \\
\end{align}

\begin{align}
\frac{dy}{dx} + 2y &= 4 \\
y(0) &= 3 \\
\mu(x)\frac{dy}{dx} + 2y\mu(x) &= 4\mu(x) \\
\mu(x) \frac{dy}{dx} + y \frac{d \mu(x)}{dx} &= \mu(x)\frac{dy}{dx} + 2y\mu(x) & \text{transforming known chain rule for } d (\mu(x)y)\text{ to given diffeq} \\
y \frac{d \mu(x)}{dx} &= 2 y \mu(x)  \\
\frac{d \mu(x)}{dx} &= 2 \mu(x)  \\
\frac{d \mu(x)}{\mu(x)} &= 2 dx \\
\int \frac{d \mu(x)}{\mu(x)} &= \int 2 dx \\
\ln{|\mu(x)|}  &= 2x + c \\
\mu(x) &= ce^{2x} & \text{solved for } \mu(x) \\
\mu(x)\frac{dy}{dx} + 2y\mu(x) &= 4\mu(x) & \text{original equation in step 2}\\
ce^{2x}\frac{dy}{dx} + 2yce^{2x} &= 4ce^{2x} & \text{substitute } \mu(x)\\
e^{2x}\frac{dy}{dx} + 2ye^{2x} &= 4e^{2x} & \text{drop }c\\
e^{2x}\frac{dy}{dx} + 2ye^{2x} &= \frac{d(ye^{2x})}{dx} & u=y, v=e^{2x}\\
\frac{d(yce^{2x})}{dx}  &= 4ce^{2x} \\
d(ye^{2x}) &= 4e^{2x} dx \\
ye^{2x} &= 2e^2x + c \\
y &= 2 + ce^{-2x} \\
y(0) &= 3 = 2 + c \implies c=1 \\
y&= 2 + e^{-2x} \\
\end{align}
